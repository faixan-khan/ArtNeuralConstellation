```python
class dummy:
    pass
opt = dummy()
opt.model = 'resnet50'
opt.finetune = False

opt.data_aug=None

opt.n_styles = 21
opt.n_genres = 18
opt.n_artists = 67
```

```python
import torch
from torchvision.models import resnet50
from torchvision import models
import torch.nn as nn

final_layers = {'resnet18':2048, 'resnet50':2048, 'resnet101':2048, 'vgg16':4096}

class Classifiers(nn.Module):
    def __init__(self, pretrained):
        super().__init__()
        backbone = getattr(models, opt.model)(pretrained=pretrained)
#         modules = list(backbone.children())
        modules = list(backbone.children())[:-1]
        self.model = nn.Sequential(*modules)       
        for param in self.model.parameters():
            param.requires_grad = False
        self.linear = nn.Sequential(nn.Linear(final_layers[opt.model], opt.n_styles))
        if opt.finetune:
            self.finetune()
        
    def finetune(self):
        for p in self.parameters():
            p.requires_grad = True
        
    
    def forward(self, ip):
        out = self.model(ip).squeeze()
        return self.linear(out)

    
class FeatExtractor(nn.Module):
    def __init__(self, model):
        super().__init__()
        modules=list(model.children())
        self.model = nn.Sequential(*modules)
    def forward(self, ip):
        return self.model(ip).squeeze()
```

```python
model = Classifiers(pretrained=True)
model.load_state_dict(torch.load(f'{opt.model}_pretrained.pth'))
```

```python
generations = {
    'StyleGAN2': './datasets/features/generated_artworks/StyleGAN2/',
    'StyleCAN2': './datasets/features/generated_artworks/StyleCAN2/',
    'StyleCWAN1': './datasets/features/generated_artworks/StyleCWAN1/',
    'StyleCWAN2': './datasets/features/generated_artworks/StyleCWAN2/',
}
```

```python
from torch.utils.data import Dataset, DataLoader, SequentialSampler
import torchvision.transforms as transforms
from PIL import Image
import glob 
import torch

class ImageFolderDataset(Dataset):
    
    def __init__(self, path, stats=None):
        self.path = path
        self.items = glob.glob(path + '*')
        self.items = [i for i in self.items if 'lowest_shape_entropy' not in i]
        if stats is None:
            self.stats = ((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
        else:
            self.stats = stats
        
    def __len__(self):
        return len(self.items)
    
    def __getitem__(self, i):
        img = Image.open(self.items[i]).convert('RGB')
        file = self.items[i]
        self.transform = transforms.Compose([transforms.ToTensor(),
                                             transforms.Normalize(*self.stats)
                                            ])
        img = self.transform(img)
        return {'img':img,
               'file': file}
```

```python
import pickle
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

def compute_features(disc, dataloader, **disc_kwargs):
    features = []
    files = []
    for x in dataloader:
        imgs = x['img'].cuda();
        file = x['file']
        with torch.no_grad():
            feats = model(imgs)
        print(f"Processed: {len(features) * 16}/{len(dataloader)*16}" , end='\r')
        features.append(feats.cpu().numpy())
        files.append(file)
    return np.concatenate(features, axis=0)
```

```python
fe = FeatExtractor(model).cuda()
```

```python
m = nn.Softmax(dim=1)
```

```python
x_axis_labels = ['early-renaissance','high-renaissance','mannerism-late-renaissance',
'northern-renaissance','baroque','rococo',
'romanticism','impressionism','pointillism+post-impressionism',
'contemporary-realism+realism+new-realism',
'art-nouveau-modern','analytical-cubism+cubism+synthetic-cubism',
'expressionism','symbolism','fauvism','action-painting+abstract-expressionism',
'color-field-painting','minimalism','na-ve-art-primitivism',
'ukiyo-e','pop-art']
y_axis_labels = x_axis_labels
```

```python
for k, g in generations.items():
    all_feats = torch.zeros(400,21)
    generated_dataset = ImageFolderDataset(g)
    dataloader = DataLoader(generated_dataset, batch_size=16, num_workers=1, sampler=SequentialSampler(generated_dataset))    
    all_feats = compute_features(fe, dataloader)
    soft_feats = m(torch.from_numpy(all_feats))
    corr = np.dot(all_feats.T,all_feats)
    corr_soft = np.dot(soft_feats.T,soft_feats)
    row_sums = corr_soft.sum(axis=1)
    eps=1e-10
    row_sums+=eps
    new_matrix = corr_soft / row_sums[:, np.newaxis]
    mask = np.zeros_like(corr)
    idx=np.argwhere(new_matrix < 0.1)
    for i in idx:
        mask[i[0]][i[1]]=1
    
    fig, ax = plt.subplots(figsize=(15,15))
    ax = plt.axes()
    sns.heatmap(new_matrix, xticklabels=x_axis_labels,
                yticklabels=y_axis_labels, mask=mask,
                annot=True, cmap=sns.cm.rocket_r,fmt='.2f',ax = ax)
    ax.set_title(k)
    plt.show()
```
